# Activation/Transfer Function


```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, FloatSlider

def plot_activation_function(x, y, title, formula_text):
    plt.figure(figsize=(8, 4))
    plt.plot(x, y, label=title, color='blue', linewidth=2)
    plt.title(f"{title}\n{formula_text}", fontsize=14)
    plt.xlabel("Input (x)")
    plt.ylabel("Output (f(x))")
    plt.axhline(0, color='black', linewidth=1, linestyle='--')
    plt.axvline(0, color='black', linewidth=1, linestyle='--')
    plt.grid(True, alpha=0.3)
    plt.show()
```

## Gaussian Error Linear Unit (GELU) Activation Function


```python
import scipy.special

def plot_gelu(scale=1.0):
    x = np.linspace(-5, 5, 200)
    y = 0.5 * (x * scale) * (1 + scipy.special.erf((x * scale) / np.sqrt(2)))
    
    formula = rf"$y = \frac{{1}}{{2}} ({scale:.1f}x)\left(1 + \mathrm{{erf}}\left(\frac{{{scale:.1f}x}}{{\sqrt{{2}}}}\right)\right)$"
    plot_activation_function(x, y, "Gaussian Error Linear Unit (GELU) Activation Function", formula)

interact(plot_gelu, scale=FloatSlider(min=0.1, max=3.0, step=0.1, value=1.0, description="Scale"));
```

### Commentary

Use Cases: Mostly in modern transformers like a BERT, GPT2, GPT3 etc.

Advantages: Smooth curve that allows better gradient pass than ReLU.

Disadvantages: Computationally more demanding. Less intuitive to tune parameters than simpler functions.


## Softplus Activation Function


```python
def plot_softplus(beta=1.0):
    x = np.linspace(-5, 5, 200)
    y = np.log(1 + np.exp(beta * x)) / beta
    
    formula = rf"$y = \frac{{1}}{{{beta:.1f}}}\,\ln\!\left(1 + e^{{{beta:.1f}x}}\right)$"
    plot_activation_function(x, y, "Softplus Activation Function", formula)

interact(plot_softplus, beta=FloatSlider(min=0.5, max=10.0, step=0.5, value=1.0, description="Beta"));
```

### Commentary

Use Cases: Modeling outputs that must always be positive. Smooth replacement for ReLU in deep neural networks.

Advantages: It is everywhere differentiable. It has always non-zero gradient.

Disadvantages: It includes logarithmic and exponential functions, which are more computationally demanding.


## Sigmoid Linear Unit (SiLU) Activation Function


```python
def plot_silu(beta=1.0):
    x = np.linspace(-5, 5, 200)
    y = x / (1 + np.exp(-beta * x))
    
    formula = rf"$y = \frac{{x}}{{1 + e^{{-{beta:.1f}x}}}}$"
    plot_activation_function(x, y, "Sigmoid Linear Unit (SiLU) Activation Function", formula)

interact(plot_silu, beta=FloatSlider(min=0.1, max=5.0, step=0.1, value=1.0, description="Beta"));
```

### Commentary

Use Cases: Usually in convolutional neural networks.

Advantages: Better accuracy than ReLU while maintaining similar stability.

Disadvantages: Computationally more demanding than ReLU.


## Gaussian Activation Function


```python
def plot_gaussian(mu=0.0, sigma=1.0):
    x = np.linspace(-5, 5, 200)
    y = np.exp(-0.5 * ((x - mu) / sigma)**2)
    
    formula = rf"$y = e^{{-\frac{{(x - {mu:.1f})^2}}{{2 \cdot {sigma:.1f}^2}}}}$"
    plot_activation_function(x, y, "Gaussian Activation Function", formula)

interact(plot_gaussian, 
         mu=FloatSlider(min=-2.0, max=2.0, step=0.5, value=0.0, description="Mean (µ)"),
         sigma=FloatSlider(min=0.2, max=2.0, step=0.1, value=1.0, description="Sigma (σ)"));
```

### Commentary

Use Cases: Especially in experimental models focused on anomaly detection or specific data clustering.

Advantages: Extremely strong response to values near the center and rapid attenuation of extremes.

Disadvantages: Not suitable for deep feed-forward networks, as information is quickly lost as it passes through the layers.


## Sinusoid Activation Function


```python
def plot_sinusoid(omega=1.0, amplitude=1.0):
    x = np.linspace(-5, 5, 200)
    y = amplitude * np.sin(omega * x)
    
    formula = rf"$y = {amplitude:.1f} \cdot \sin({omega:.1f}x)$"
    plot_activation_function(x, y, "Sinusoid Activation Function", formula)

interact(plot_sinusoid, 
         omega=FloatSlider(min=0.5, max=5.0, step=0.5, value=1.0, description="Omega (ω)"),
         amplitude=FloatSlider(min=0.5, max=3.0, step=0.5, value=1.0, description="Amplitude"));
```

### Commentary

Use Cases: In processing periodic signals and time series with clear seasonality.

Advantages: It can model fine details and periodic patterns that other functions cannot capture.

Disadvantages: They are not monotonic and can complicate gradient learning.
